EXO 1 : vous disposer d'une base de donner contenant le catalogue de produit d'une boutique
question : quel sont les 10 produits les mieux note de ce catalogue ? 

pour repondre a cette question on peut proceder comme suite : 

etape : afficher la base de donne 


```python
SELECT *
FROM products;
```

la requette precedengte permet de visualiser de maniere total la table product  

etape 2 selectionner les champs concerner c'esxt a dire ceux qui permettent d'afficher 
uniquement le produit ainsis que le chanp qui represente sa note afin de mieux visualiser 

```python
SELECT id,name,rating
FROM products;
```

cete requette affiche uniquement l'id du produit sont nom et sa note  

etape 3 afficher les prodit par order de note decroissante c'est a dire du mieux noter au moin noter 

```python
SELECT id,name,rating
FROM products
ORDER BY rating DESC;
```

etape 5 limiter le nombre de produits afficher au top 10 

```python
SELECT id,name,rating
FROM products
ORDER BY rating DESC LIMIT (10);
```
